```python
# chama biblioteca
from llama_parse import LlamaParse
from llama_index.core import SimpleDirectoryReader


def llama_parse_extract_text_from_pdf(pdf_path):
    # extração de dados de um documento PDF
    documento = LlamaParse(result_type="markdown",
                           verbose=True,
                           language='pt',
                           #disable_ocr=False,
                           # podemos especificar as páginas que queremos extrair os dados
                           #target_pages="0,2,7",
                           ## podemos especificar a área da página que queremos extrair os dados entre 0 e 1
                           ## bbox_left, bbox_right, bbox_tope, bbox_bottom
                           ## Essa configuração abaixo é para os modelos de artigos do IAC
                           bounding_box="0.1, 0.09, 0.2, 0.1",
                           ).load_data(pdf_path)

    return documento
```

```python
documento = llama_parse_extract_text_from_pdf(r'C:/Estudos e Projetos/2024-LLM-Agronomo-Virtual/pdf/0000444-08.2022.8.26.0052_relatório_final_pags_73-77.pdf')
```

```python
## salvar aquivo
nome_do_arquivo = "outro.md"

with open(nome_do_arquivo, "w") as arquivo:
    for item in documento:
        arquivo.write(item.text)
```
